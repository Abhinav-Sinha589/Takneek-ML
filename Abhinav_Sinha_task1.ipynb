```python
import numpy as np
import matplotlib.pyplot as plt

def generate_random_points(dimension, num_points, min_coord, max_coord):
    return np.random.uniform(min_coord, max_coord, size=(num_points, dimension))

def divide_into_cells(points, grid_size):
    return np.floor(points / grid_size).astype(int)

def calculate_distance(point1, point2):
    return np.linalg.norm(point1 - point2)

def find_pairs(points, r):
    pairs = set()
    valid_importpairs = set()
    num_points, dimension = points.shape
    grid_size = 2 * r
    cells = divide_into_cells(points, grid_size)
    for i in range(num_points):
        current_point = points[i]
        current_cell = cells[i]
        valid_pairs = set()
        for delta in np.ndindex(*([3]*dimension)):
            neighbor_cell = current_cell + np.array(delta) - 1
            if np.all((0 <= neighbor_cell) & (neighbor_cell < grid_size)):
                mask = np.all(cells == neighbor_cell, axis=1)
                neighbor_points = points[mask]
                for neighbor_point in neighbor_points:
                    if calculate_distance(current_point, neighbor_point) < r:
                        valid_importpairs.add((tuple(current_point), tuple(neighbor_point)))
        pairs.update(valid_pairs)

    return pairs



def plot_points(points, cells, title,grid_size):
    grid_boxes = np.floor(random_points_2d / grid_size).astype(int)
    plt.scatter(points[:, 0], points[:, 1], c=grid_boxes.sum(axis=1), cmap='viridis')  # Use a single color for all points
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

random_points_2d = np.random.uniform(0, 100, size=(1000, 2))
grid_size = 1
grid_boxes = np.floor(random_points_2d / grid_size).astype(int)


plt.scatter(random_points_2d[:, 0], random_points_2d[:, 1], c=grid_boxes.sum(axis=1), cmap='viridis')
plt.title('2D Space - Grid for r=1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()


# Task 1
dimension_2_points = generate_random_points(2, 1000, 0, 100)

# Task 2
for r in range(1, 11):
    pairs_2D = find_pairs(dimension_2_points, r)
    plot_points(dimension_2_points, divide_into_cells(dimension_2_points, 2 * r), f'2D Space - Grid for r={r}',2*r)
    print(f'Pairs for r={r}: {pairs_2D}\n')

# Task 3
dimensions = [3, 4, 10, 50, 100]
results = {}

for d in dimensions:
    points_d = generate_random_points(d, 1000, 0, 100)
    pairs_d = {r: find_pairs(points_d, r) for r in range(1, 11)}
    results[d] = pairs_d

```
